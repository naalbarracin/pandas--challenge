```python
# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"
#os.getcwd()
# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Purchasing Analysis (Total)


* Run basic calculations to obtain number of unique items, average price, etc.
* Create a summary data frame to hold the results
* Optional: give the displayed data cleaner formatting
* Display the summary data frame

```python
#create the variables to set the values
uitems = purchase_data['Item Name'].nunique()
average = purchase_data.Price.mean()
tpurchases = purchase_data['Purchase ID'].count()
trevenue = purchase_data.Price.sum()
#create a new data frame and added the colums
twoDf = pd.DataFrame()
twoDf['Number of Unique Items'] = [uitems]
twoDf['Average Price'] = [average]
twoDf['Number of Purchases'] = [tpurchases]
twoDf['Total Revenue'] = [trevenue]
twoDf
```

## Gender Demographics
* Percentage and Count of Male Players
* Percentage and Count of Female Players
* Percentage and Count of Other / Non-Disclosed

```python
d = {'SN':'Total Count'}

gcount = purchase_data.rename(columns=d).groupby('Gender').agg({'Total Count': 'nunique'})
gcount['Percentage of Players'] = purchase_data.groupby('Gender').agg({'SN': 'nunique'})/purchase_data['SN'].nunique()*100
gcount

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
* Create a summary data frame to hold the results
* Optional: give the displayed data cleaner formatting
* Display the summary data frame

```python
results = purchase_data.groupby('Gender').agg(PurchaseCount=('SN', 'count'), AveragePurchasePrice=('Price', 'mean'), TotalPurchaseValue=('Price', 'sum'),UniqueCount=('SN', 'nunique'))
results['PercentageofPlayers'] = results['TotalPurchaseValue']/results['UniqueCount']
results[['PurchaseCount', 'AveragePurchasePrice','TotalPurchaseValue','PercentageofPlayers']]
```

## Age Demographics

* Establish bins for ages
* Categorize the existing players using the age bins. Hint: use pd.cut()
* Calculate the numbers and percentages by age group
* Create a summary data frame to hold the results
* Optional: round the percentage column to two decimal points
* Display Age Demographics Table

```python
purchase_data['age_bins'] = pd.cut(x=purchase_data['Age'], bins=[0,9, 14, 19, 24, 29, 34, 39, 100])
purchase_data['age_range_bins'] = pd.cut(x=purchase_data['Age'], bins=[0,9, 14, 19, 24, 29, 34, 39,100], labels=['<10','10-14', '15-19', '20-24','25-29','30-34','35-39','40+'])
purchase_data_new = purchase_data.set_index(['age_range_bins'])
purchase_data_age = purchase_data_new.groupby('age_range_bins').agg({'SN': 'nunique'})

purchase_data_age['Percentage of Players'] = round(purchase_data_age.SN/purchase_data.SN.nunique()*100,2)

purchase_data_age.rename(columns={'SN': 'Total Count'}, inplace=True)
purchase_data_age
```

## Purchasing Analysis (Age)
* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data_new= purchase_data_new.groupby('age_range_bins').agg(PurchaseCount=('SN', 'count'), AveragePurchasePrice=('Price', 'mean'), TotalPurchaseValue=('Price', 'sum'),UniqueCount=('SN', 'nunique'))
purchase_data_new['PercentageofPlayers'] = purchase_data_new['TotalPurchaseValue']/purchase_data_new['UniqueCount']
purchase_data_new
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
d = {'Item ID':'Total_ID'}

top_purchase_data = purchase_data.rename(columns=d)
top_purchase_data = top_purchase_data.groupby('SN').agg(PurchaseCount=('Total_ID', 'count'), AveragePurchasePrice=('Price', 'mean'), TotalPurchaseValue=('Price', 'sum'))

top_purchase_data.sort_values("TotalPurchaseValue", ascending=False).head(5)
```

## Most Popular Items

*Retrieve the Item ID, Item Name, and Item Price columns

*Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value

*Create a summary data frame to hold the results

*Sort the purchase count column in descending order

*Optional: give the displayed data cleaner formatting

*Display a preview of the summary data frame

```python
d = {'Item ID':'Item_ID','Item Name':'Item_Name'}

most_purchase_data = purchase_data.rename(columns=d)
most_purchase_data= most_purchase_data.groupby(['Item_ID','Item_Name']).agg(PurchaseCount=('Item_ID', 'count'), AveragePurchasePrice=('Price', 'mean'),TotalPurchaseValue=('Price', 'sum'))
#most_purchase_data

most_purchase_data.sort_values("PurchaseCount", ascending=False).head(5)
```
